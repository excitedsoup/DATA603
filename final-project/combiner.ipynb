```python
import pandas as pd
```

```python
weatherdf = pd.read_csv('2013_2019.csv')
crashdf = pd.read_csv('cleaned_crashes.csv')
```

```python
weatherdf = weatherdf.drop(columns = weatherdf.columns[0])
crashdf = crashdf.drop(columns = crashdf.columns[0])
```

```python
combined = weatherdf.join(crashdf)
```

```python
combined

combined = combined.drop(columns = ['DATE', 'ReportDate'])
```
